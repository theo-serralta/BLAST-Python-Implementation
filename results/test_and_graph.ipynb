Multithreading Time Comparison Graph

```python
import matplotlib.pyplot as plt

# Data for 1000 and 2000 sequences
cores = ['1c', '2c', '3c', '4c', '5c', '6c', '7c', '8c', 'blast']
times = [233, 123, 98, 86, 83, 84, 81, 84, 0.2]
times_2000_seq = [434, 251, 176, 173, 168, 143, 143, 147, 0.3]

# Creating a side-by-side bar chart for 1000 and 2000 sequences
bar_width = 0.35  # Width of the bars
index = range(len(cores))  # Index for bar placement

plt.figure(figsize=(8, 6))

# Bars for 1000 sequences
plt.bar(index, times, bar_width, color='darkgray', edgecolor='black', label='1000 sequences')

# Bars for 2000 sequences
plt.bar([i + bar_width for i in index], times_2000_seq, bar_width, color='lightgray', edgecolor='black', label='2000 sequences')

# Adding labels and a title
plt.title('Execution time vs. Number of Cores (1000 vs 2000 sequences)', fontsize=14)
plt.xlabel('Number of Cores', fontsize=12)
plt.ylabel('Time (seconds)', fontsize=12)

# Adding ticks on the x-axis
plt.xticks([i + bar_width / 2 for i in index], cores)

# Displaying the legend
plt.legend()

# Aesthetic adjustments for a scientific paper format
plt.grid(True, which='both', axis='y', linestyle='--', linewidth=0.7)
plt.tight_layout()

# Displaying the bar chart
plt.show()
```

```python
Mann-Whitney Test to Compare E-values
```

```python
from scipy.stats import mannwhitneyu

# E-values or bit scores from your BLAST and BLAST+
e_values_blast_python = [8.01e-165, 1.20e-24, 3.73e-18, 9.9e-14, 1.59e-11, 1.87e-09, 3.86e-6]  # values from the implementation
e_values_blast_plus = [0.00, 1e-22, 1e-16, 1e-11, 2e-09, 1e-07, 3e-04]  # values from BLAST+

# Mann-Whitney test
stat, p = mannwhitneyu(e_values_blast_python, e_values_blast_plus)
print(f'Statistic: {stat}, p-value: {p}')
```

Mann-Whitney Test to Compare Bit scores

```python
# Bit scores from your BLAST and BLAST+
bit_scores_blast_python = [573, 107.5, 85.94, 71.26, 63.92, 57.04, 46.02]  # values from the implementation
bit_scores_blast_plus = [563, 95.7, 77.7, 62.7, 55, 49.5, 39.2]  # values for BLAST+

# Mann-Whitney test
stat, p = mannwhitneyu(bit_scores_blast_python, bit_scores_blast_plus)
print(f'Statistic: {stat}, p-value: {p}')
```

Mann-Whitney test to compare identity percentages

```python
# Percentage identity from your BLAST and BLAST+
identities_blast_python = [100, 28, 24, 24, 25, 26, 24]  # values from your implementation
identities_blast_plus = [100, 28, 24, 33, 26, 27, 47]  # values from BLAST+

# Mann-Whitney test
stat, p = mannwhitneyu(identities_blast_python, identities_blast_plus)
print(f'Statistic: {stat}, p-value: {p}')
```
